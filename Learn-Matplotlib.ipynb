```python
from matplotlib import pyplot as plt
import numpy as np
```

**line plot**

```python
x=np.arange(1,11)
y=2*x
```

```python
plt.plot(x,y)
plt.title("line plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
plt.plot(x,y,color='red',linestyle=":",linewidth=2)
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='red',linestyle=":",linewidth=2)
plt.plot(x,y2,color='green',linestyle="-",linewidth=3)
plt.title("line plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True)
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='red',linestyle=":",linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y2,color='green',linestyle="-",linewidth=3)

plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color='red',linestyle=":",linewidth=2)

plt.subplot(2,1,2)
plt.plot(x,y2,color='green',linestyle="-",linewidth=3)

plt.show()
```

**bar plot**

```python
stud={"a":87,"b":56,"c":98}
name=list(stud.keys())
value=list(stud.values())
```

```python
plt.bar(name,value)
plt.title("bar plot")
plt.xlabel("names")
plt.ylabel("values")
plt.show()
```

```python
plt.barh(name,value,color='g')
plt.title("bar plot")
plt.xlabel("names")
plt.ylabel("values")
plt.grid(True)
plt.show()
```

**scatter plot**

```python
x=[10,20,30,40,50,60,70,80,90]
y=[8,6,1,2,9,3,5,6,0]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,marker="*",color="green",s=200) #s is size of the point 
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
y=[8,6,1,2,9,3,5,6,0]
y2=[2,4,10,7,3,0,9,1,5]
```

```python
plt.scatter(x,y,marker="*",color="green",s=200) 
plt.scatter(x,y2,marker=".",color="red",s=100) 
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y,marker="*",color="green",s=200) 
plt.subplot(1,2,2)
plt.scatter(x,y2,marker=".",color="red",s=100) 
plt.show()
```

```python
plt.subplot(2,1,1)
plt.scatter(x,y,marker="*",color="green",s=200) 
plt.subplot(2,1,2)
plt.scatter(x,y2,marker=".",color="red",s=100) 
plt.show()
```

**histogram**

**box plot**

```python
one=[1,2,3,4,5,6,7,8,9]
two=[6,2,5,3,4,6,5,3,3]
three=[6,3,5,7,2,3,4,5,2]
data=list([one,two,three])
```

```python
plt.boxplot(data)
plt.show()
```

**violin plot**

```python
one=[1,2,3,4,5,6,7,8,9]
two=[6,2,5,3,4,6,5,3,3]
three=[6,3,5,7,2,3,4,5,2]
data=list([one,two,three])
```

```python
plt.violinplot(data,showmedians=True)
plt.show()
```

**pie chart**

```python
fruit=["apple","orange","mango","grapes"]
qual=[31,72,43,87]
```

```python
plt.pie(qual,labels=fruit)
plt.show()
```

```python
plt.pie(qual,labels=fruit,autopct="%0.1f%%",colors=["red","orange","yellow","green"])
plt.show()
```

```python
plt.pie(qual,labels=fruit,radius=1)
plt.pie([1],colors="w",radius=0.5)
plt.show()
```
